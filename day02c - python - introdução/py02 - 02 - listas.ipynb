## Usando linspace: dividindo um intervalo em n pontos, incluso o primeiro e o último

- Logo, se vc criou uma lista de 1 até 20 - você terá, por exemplo, 19 divisições internas +1 inicio e +1 o fim = 21 pontos

````python
mat3 = np.arange(1, 21)
mat3
````

````python
# ou da mesma forma
mat3 = np.arange(1, 21, step=1)
mat3
````

````python
# ou com linspace - 19+2 pontos
mat3 = np.linspace(1, 21, 21)
mat3
````

````python
# iniciar de zero tem que aumentar +1
mat3 = np.linspace(0, 21, 22)
mat3
````

## Como você pode obsevar uma LISTA tem um comportamento diferente que uma Lista em Numpy

## Você acha que o numpy é rápido

 - %%time é um meta-comando --> pede para calcular o tempo transcorrido para processar a célula

````python
%%time

total=0

lista = np.arange(0, 1000*1000*10)
for i in lista:
    # somando 1 em total --> igual a total = total + 1
    total += i
    
print("---- fim:", total)    
````

````python
%%time

lista = np.arange(0, 1000*1000*10)
total = np.sum(lista)
    
print("---- fim:", total)
````

## Porque o numpy é ao menos 60 vezes mais rápido?
