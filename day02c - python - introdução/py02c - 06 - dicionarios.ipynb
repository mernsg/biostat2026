## Outros tipos de chaves



```python
dk = {}
dk['celula'] = 'nm'
dk['alfinete'] = 'mm'
dk['mesa'] = 'm'
dk['estrada'] = 'km'

len(dk), type(dk)
```


## Como vejo as chaves? dict.keys() --> chaves



```python
dk.keys()
```

## Como vejo os valores? dict.values() --> valores



```python
dk.values()
```

## Como crio um iterator? Há duas formas
  - dict.keys()
  - dict.items()

```python
for key in dk.keys():
  print(key)
```

### Quero ver chave e valor

```python
for key in dk.keys():
  print(key, dk[key])
```

```python
### mais bonito !!

for key in dk.keys():
  print("chave: '%s' tem valor: '%s'"%(key, dk[key]))
```

```python
### mais bonito !! com tabulador '\t'

for key in dk.keys():
  print("chave: '%s' \ttem valor: '%s'"%(key, dk[key]))
```

## E via items()
  - items() devolver chave e valor

```python
for key, val in dk.items():
  print("chave: '%s' tem valor: '%s'"%(key, val))
```

### Comprehension list

```python
lista = [1, 3, 5, 7, 9]
lista2 = [i**2 for i in lista]

lista2, type(lista2), len(lista2)
```

```python
lista2 = [i**2 for i in range(5)]

lista2, type(lista2), len(lista2)
```

## Criar um dicionário com comprehension

  - comprehension entre chaves {}

```python
dr = {i: i**2 for i in range(5)}
len(dr), type(dr)
```

```python
for key, val in dr.items():
  print(f"valor {val:3} é o quadrado de {key:2}")
```
