## Sets: sets são listas sem repetição

```python
lista = [4, 5, 8, 2, 4, 3, 5, 9, 12]
print(len(lista), type(lista))

# criando s1 com dados da lista
s1 = set( lista )
print(len(s1), type(s1))
```

```python
s1
```

## O que mudou?

## Set é muito restrito, você não consegue filtrar elemetos como em list

```python
s1[0]
```

```python
s1[2:4]
```

## Mas, você consgue iterar

```python
for elem in s1:
  print(elem)
```

## Você pode voltar um set para lista e aí tem todas as funções

```python
s2 = list(s1)
len(s2), type(s2), s2
```

## Como consigo mitetizar um Set com lista?

```python
lista = [4, 5, 8, 2, 4, 3, 5, 9, 12]
print(len(lista), type(lista))
```

```python
import numpy as np
```

```python
lista2 = np.unique(lista)
print(len(lista2), type(lista2))
```

```python
lista2
```

```python
list(lista2) == list(s2)
```

## O set é uma lista em repetição ordenada
### lista.sort() - ordena uma lista
### np.unique(lista) - ordena e tira repetidos
