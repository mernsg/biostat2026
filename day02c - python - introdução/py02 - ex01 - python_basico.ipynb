### Posso definir limites para o plot

```python
fig = plt.figure(figsize = (12,8))

plt.plot(xs, ys, color='green', linewidth=2)
plt.title('função quadrática')
plt.xlabel("seq x")
plt.ylim(0,400)
plt.ylabel("y = x**2");
```

### Posso coloar uma linha vertical em x = 0

```python
fig = plt.figure(figsize = (12,8))

plt.plot(xs, ys, color='green', linewidth=2)
plt.title('função quadrática')
plt.xlabel("seq x")
plt.ylim(0,400)
plt.xlim(-20,20)
plt.axvline(x = 0, color = 'darkred')
plt.ylabel("y = x**2");
```
