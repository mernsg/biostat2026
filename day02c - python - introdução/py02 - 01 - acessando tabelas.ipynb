### Se não conseguir ler: pare

```python
# 'r' == read only

with open(fname, 'r') as f:  # , encoding='utf-8'

    while True:
        try:
            # leia uma linha
            line = f.readline()

            # se linha vazia então break (cai fora do loop)
            if line == '':
                break
            print(line)
        except:
            break

#-- \n == carriage return
print("\n---- fim ----")
```

### Numerar linhas

```python
num = 0

with open(fname, 'r') as f:  # , encoding='utf-8'

    while True:
        try:
            line = f.readline()

            if line == '':
                break

            num += 1
            print(f"{num})", line)
        except:
            break

print("\n---- fim ----")
```

### Processando arquivo texto
### Abrindo em uma matriz com 2 termos

\<string>.strip() - remova brancos antes de depois da string  
\<string>.split(pattern) - quebre o string de acordo com um pattern (modelo, substring)

```python
fname, os.path.exists(fname)
```

```python
with open(fname, 'r') as f:  # , encoding='utf-8'

    while True:
        try:
            line = f.readline()
            if line == '':
                break

            line = line.strip() # tire os espaços antes e depois
            mat = line.split('-')
            print(mat)
        except:
            break

#-- \n == carriage return
print("\n---- fim ----")
```

### Processando arquivo texto
### Abrindo em uma matriz com 3 termos

```python
with open(fname, 'r') as f:  # , encoding='utf-8'

    while True:
        line = f.readline()
        if line == '':
            break

        line = line.strip() # tire os espaços antes e depois
        mat = line.split('-')

        termo0 = mat[0]
        mat = mat[1].split(' ')
        termo1 = mat[0]
        termo2 = mat[1]

        print(termo0, '\t', termo1, '\t', termo2)


#-- \n == carriage return
print("\n---- fim ----")
```

### Formatando

```python
with open(fname, 'r') as f:  # , encoding='utf-8'

    while True:
        line = f.readline()
        if line == '':
            break

        line = line.strip() # tire os espaços antes e depois
        mat = line.split('-')

        termo0 = mat[0]
        mat = mat[1].split(' ')
        termo1 = mat[0]
        termo2 = mat[1]

        print(f"{termo0:12} \t {termo1:8} \t {termo2}")


#-- \n == carriage return
print("\n---- fim ----")
```
