## Formas de armazenamento de dados em memória

## Dicionário:
  - dicionários são MUITO usados em python
  - são elementos chave: valor
  - dada uma chave --> ele devolver o valor

```python
# definindo um dictionary com dict()
dm = dict()
```

```python
# dm é um dictionary
len(dm), type(dm)
```

## Como coloco elementos no dicionário
  - há várias formas, vamos ver 2 delas


```python
dm[0] = 'elem0'
dm[1] = 'elem1'
dm[2] = 'elem2'
dm
```

```python
# qual o comprimento de dm?
len(dm)
```

### Chave - valor
  - dada uma chave --> devolver valor
  - dm[0] --> devolve elem0

```python
dm[0]
```

```python
# percorrendo todos elementos
for i in range(len(dm)):
  print(i, dm[i])
```

## Outros tipos de chaves



```python
dk = {}
dk['celula'] = 'nm'
dk['alfinete'] = 'mm'
dk['mesa'] = 'm'
dk['estrada'] = 'km'

len(dk), type(dk)
```


## Como vejo as chaves? dict.keys() --> chaves



```python
dk.keys()
```

## Como vejo os valores? dict.values() --> valores



```python
dk.values()
```

## Como crio um iterator? Há duas formas
  - dict.keys()
  - dict.items()

```python
for key in dk.keys():
  print(key)
```

### Quero ver chave e valor

```python
for key in dk.keys():
  print(key, dk[key])
```

```python
### mais bonito !!

for key in dk.keys():
  print("chave: '%s' tem valor: '%s'"%(key, dk[key]))
```

```python
### mais bonito !! com tabulador '\t'

for key in dk.keys():
  print("chave: '%s' \ttem valor: '%s'"%(key, dk[key]))
```

## E via items()
  - items() devolver chave e valor

```python
for key, val in dk.items():
  print("chave: '%s' tem valor: '%s'"%(key, val))
```

## Outra forma de criar um dicionário:
  - via comprehension
    - comprehension entre chaves {}

```python
dr = {i: i**2 for i in range(5)}
len(dr), type(dr)
```

```python
for key, val in dr.items():
  print("valor %d é o quadrado de %d"%(val, key))
```
