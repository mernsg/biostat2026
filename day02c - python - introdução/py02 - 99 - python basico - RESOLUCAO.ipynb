### Vamos criar uma função quadrática, para tanto

1. criar uma lista_x
2. gerar o quadrado da lista_x e por em lista_y
3. plotar no matplotlib

```python
xs = np.arange(0, 100)
# os 10 primeiros elementos
xs[:10]
```

```python
# vamos dividir cada elemento pelo valor 5
xs = xs/5
# repare que o 0 é optativo: xs[0:10] == xs[:10]
xs[:10]
```

```python
# vamos gerar o valor quadrático
ys = xs * xs
ys[:10]
```

### Plotando com o matplotlib

```python
fig = plt.figure(figsize = (12,8))

plt.plot(xs, ys, color='green', linewidth=2)
plt.title('função quadrática')
plt.xlabel("seq x")
plt.ylabel("y = x**2");
```

### Posso coloar uma linha vertical em x = 0

```python
fig = plt.figure(figsize = (12,8))

plt.plot(xs, ys, color='green', linewidth=2)
plt.title('função quadrática')
plt.xlabel("seq x")
plt.ylim(0,25)
plt.axvline(x = 0, color = 'darkred')
plt.ylabel("y = x**2");
```
