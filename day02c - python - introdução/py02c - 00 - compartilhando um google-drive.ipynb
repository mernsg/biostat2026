```python
from google.colab import auth
import gspread
from google.auth import default
import pandas as pd

from gspread_dataframe import get_as_dataframe
```

### Autenticando o Usuário para poder ler o google-drive

```python
auth.authenticate_user()

creds, _ = default()
gc = gspread.authorize(creds)
```

### O diretório de dados (data) confere

```python
root_data, os.path.exists(root_data)
```

### gc.open localiza a tabela Iris gsheet (nome tem que ser único)

```python
sh = gc.open('Iris')
sh
```

```python
for ws in sh.worksheets():
    print(ws.title, ws.row_count, ws.col_count)
```

```python
worksheet = sh.worksheet("Sheet1")

df = get_as_dataframe(worksheet)
df = df.dropna(how="all")  # remove empty rows
print(">>> lines", len(df), '\n')
df.head()
```
