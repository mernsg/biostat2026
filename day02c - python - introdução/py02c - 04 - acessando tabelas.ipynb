### Lendo um arquivo ASCII (letras, não binário)

```python
# 'r' == read only

with open(filename, 'r') as f:  # , encoding='utf-8'

    while True:
        try:
            # leia uma linha
            line = f.readline()

            # se linha vazia então break (cai fora do loop)
            if line == '':
                break
            print(line)
        except:
            break

#-- \n == carriage return
print("\n---- fim ----")
```

### Numerar linhas e tirar \n

```python
num = 0

with open(filename, 'r') as f:  # , encoding='utf-8'

    while True:
        try:
            line = f.readline()

            if line == '':
                break

            # strip() limpe espaços
            # line[:-1] tire o ultimo caracter = '\n' - carriage return
            line = line.strip()

            num += 1
            # num = num + 1
            print(f"{num}) {line}")
        except:
            break

print("\n---- fim ----")
```

### Processando arquivo texto
### Abrindo em uma matriz com 2 termos

\<string>.strip() - remova brancos antes de depois da string  
\<string>.split(pattern) - quebre o string de acordo com um pattern (modelo, substring)

```python
stri = 'aaaa-bbbb-cccc'
async
```

```python
num = 0

with open(filename, 'r') as f:  # , encoding='utf-8'

    while True:
        try:
            line = f.readline()

            if line == '':
                break

            # strip() limpe espaços
            # line[:-1] tire o ultimo caracter = '\n'
            line = line.strip()

            mat = line.split('-')

            num += 1
            print(f"{num}) {mat[0]:8} - {mat[1]}")
        except:
            break

print("\n---- fim ----")
```

### Processando arquivo texto
### Abrindo em uma matriz com 3 termos

```python
num = 0

with open(filename, 'r') as f:  # , encoding='utf-8'

    while True:
        try:
            line = f.readline()

            if line == '':
                break

            # strip() limpe espaços
            # line[:-1] tire o ultimo caracter = '\n'
            line = line.strip()
            mat = line.split('-')
            mat2 = mat[1].split(' ')

            num += 1
            print(f"{num}) {mat[0]:7} - {mat2[0]:5} - {mat2[1]}")
        except:
            break

print("\n---- fim ----")
```
